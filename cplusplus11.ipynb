## Funciones Lambda

```c++
printf("\n////////////////// Samples by value");
std::string nombre = "Alex";

auto myFunc = []() { printf("\n Hola"); };
auto myFuncCapturing = [nombre]() { printf("\n Hola %s", nombre.c_str()); };
auto myFuncPassing = [](const char* nombre1) { printf("\n Hola %s", nombre1); };
auto myFuncCapturingAll = [=]() { printf("\n Hola %s", nombre.c_str()); };

myFunc();
myFuncCapturing();
myFuncPassing("Alex2");
myFuncCapturingAll();
                     
printf("\n////////////////// Samples by reference");
std::string nombre1ByRef = "Alex";
std::string nombre2ByRef = "Hermes";

auto myFuncCapturingReference = [&nombre1ByRef, &nombre2ByRef]() { printf("\n Hola %s %s", nombre1ByRef.c_str(), nombre2ByRef.c_str()); };
auto myFuncCapturingReferenceAll = [&]() { printf("\n Hola %s", nombre.c_str()); };

myFuncCapturingReference();
myFuncCapturingReferenceAll();
                                                                  
nombre1ByRef = "Alex2";
nombre = "Juan";
myFuncCapturingReference();
myFuncCapturingReferenceAll();
```

Entendemos que los warnings son debidos a usar string. Si usaramos tipos primitivos se pasarían por valor.
